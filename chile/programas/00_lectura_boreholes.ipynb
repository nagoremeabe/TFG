## 02 - Mapa de los *Boreholes*
Obtener un mapa de las posiciones de los pozos y donde aparezca información adicional como su profundidad, coordenadas geográficas,...etc.

```python
# Librerías para graficas los boreholes en un mapa
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
# El primer valor pertenece al borehole "ala1110-2"
# El segundo valor al "1501"

lat=[-28.66632,-26.75389]
lon=[-70.9104,-69.895]
depth=[412.0,398.37]
```

```python
# Plot de la información
fig = plt.figure(figsize=(8,8))
m = Basemap(projection="lcc",resolution="h",lat_0= -28,lon_0=-70,width=2E6,height=2.2E6)
m.shadedrelief()
m.drawcoastlines(color ="gray")
m.drawcountries(color="gray")
m.drawstates(color="gray")
m.scatter(lon,lat,latlon=True,c = depth ,cmap = "Reds",marker="o",s=100,edgecolor="k")

x,y=m(lon[1],lat[1])
plt.annotate(f"""Nombre: 1501
Profundidad: {depth[1]} m
Latitud: {lat[1]}S
Longitud: {lon[1]}W""",
            xy=(x,y),
            xytext=m(lon[1], lat[1]+1),
            arrowprops=dict(arrowstyle="fancy", color='black'),
            backgroundcolor="antiquewhite")

x,y=m(lon[0],lat[0])
plt.annotate(f"""Nombre: ala1110-2
Profundidad: {depth[0]} m
Latitud: {lat[0]}S
Longitud: {lon[0]}W""",
            xy=(x,y),
            xytext=m(lon[0], lat[0]-4),
            arrowprops=dict(arrowstyle="fancy",color='black'),
            backgroundcolor="antiquewhite")

bar = plt.colorbar()
bar.set_label("\nProfundidades (m)")
plt.title("Posición de los Boreholes en Chile y sus Profundidades.")
```
