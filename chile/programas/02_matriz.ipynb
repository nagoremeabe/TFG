## 04 - Calculo de la Matriz $A_{jk}$

Se obtendrá a continuación la matriz definida mediante esta expresión matemática para cada uno de los *boreholes* y se guardará en un ".csv".

$A_{jk} = erfc\left\lbrace \frac{z_j}{2 \sqrt{kt_{k-1}}}\right\rbrace - erfc\left\lbrace \frac{z_j}{2 \sqrt{kt_{k}}}\right\rbrace
$

```python
# Librerías para el tratamiento de datos
import numpy as np
import pandas as pd

# Librerías para la visulización de los datos
import matplotlib.pyplot as plt
import seaborn as sns

# Otras librerías importantes
import math
import warnings
warnings.filterwarnings("ignore")
```

```python
# Determinar el estilo y los colores
sns.set_theme(style="whitegrid")
sns.color_palette('Set2')
sns.color_palette("crest", as_cmap=True)
```

```python
# Parámetros elegidos
tk = np.arange(0,520,20)
alpha = 31.6 #m**2*y**-1
```

```python
# Función para calcular la cantidad de datos que tenemos antes de una profundidad determinada
#============================================================================================
def count(profundidad,zj):
    suma = 0
    for l in zj:
        if (l<=profundidad):
            suma += 1
    return suma
```

### 04.1 - ala1110-2

```python
nombre = "ala1110-2"
path = "../boreholes/ala1110-2.dat"
zj,Tj = np.loadtxt (path , skiprows = 5,unpack = True)
```

```python
# Llenar la matriz de 0-s.
Ajk = []
filas = len(zj)
columnas = len(tk)

for i in range(filas):
    Ajk.append([])
    for j in range(columnas):
        Ajk[i].append(0)

#Rellenar la matriz con los valores necesarios        
for j in range(filas):
    for k in range(columnas):
        Ajk[j][k] = -math.erfc(zj[j] / (2 * math.sqrt(alpha * tk[k-1]))) + math.erfc(zj[j] / (2 * math.sqrt(alpha*tk[k])))
        
```

```python
# Dimensiones de la matriz: j x k
j = len(Ajk)
k = len(Ajk[0])

# Plot de la matriz Ajk
fig = plt.figure()
ax = fig.add_subplot(111)
pos =ax.imshow(Ajk,interpolation="nearest",cmap="crest")
ax.set_aspect('equal')
ax.set_aspect('auto')
plt.title("Plot del Kernel de Datos: Ajk")
plt.gca().invert_yaxis()

ax.set_ylabel("Índice de filas")
ax.set_xlabel("Índice de columnas")
fig.colorbar(pos, ax=ax)
```

```python
#Quito una columna por que da mal siempre (una k)
Ail = np.delete(Ajk,0,axis=1)

#Las filas se quita en profundidades menores a 30m (resta de j-s)
#Ail = Ail [count(30,zj):] 

# Dimensiones de la nueva matriz
i = len(Ail)
l = len(Ail[0])

# Plot de la matriz Ail
fig = plt.figure()
ax = fig.add_subplot(111)
pos =ax.imshow(Ail,interpolation="nearest",cmap="crest")
ax.set_aspect('equal')
ax.set_aspect('auto')
plt.title("Plot del Kernel de Datos: Ail")
plt.gca().invert_yaxis()

ax.set_ylabel("Índice de filas")
ax.set_xlabel("Índice de columnas")
fig.colorbar(pos, ax=ax)
```

```python
# Guardar los datos de la matriz Ail
nombre = "ala1110-2"
pd.DataFrame(Ail).to_csv(f"../resultados/matriz_{nombre}.csv",index=False)
```
