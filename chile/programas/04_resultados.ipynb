## 06 - Visualización y Comparación de los Resultados

```python
# Librerías para el tratamiento de datos
import numpy as np
import pandas as pd

# Librerías para la visulización de datos
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
```

### 06.1 - ala1110-2

```python
# Leer datos medidos y estimados y definir parámetros interesante
nombre="ala1110-2"
datak = list(range(1880,2021))
datak_interesantes = list(range(1951,2021))

df_sat=pd.read_csv("../resultados/sat.csv",index_col="Fechas")
mest=np.genfromtxt(f"../resultados/mest_{nombre}.csv",delimiter=",")
```

```python
# Plot de las anomalías en el aire
plt.figure(figsize=(15,5))
plt.plot(df_sat.index,df_sat[nombre])
plt.xticks(ticks=df_sat.index[::5],rotation=90)
plt.title(f"Anomalías de Temperatura {nombre} con Media Anual")
```

```python
# "Normalización" de las anomalías en el aire 
datak = list(range(1880,2021))
To = df_sat[datak.index(1980):datak.index(2009)]

To =list(To.mean(axis=0))
# El primero pertenece al borehole ala el otro a 1501
df_sat["ala1110-2"]
anom = df_sat["ala1110-2"]-To[0]
anom_1501= df_sat["1501"]-To[1]

#Plot de anomalías registradas en el aire normalizadas y no normalizadas
nombre="ala1110-2"
plt.figure(figsize=(15,5))
plt.plot(df_sat.index,df_sat[nombre],label="Datos")
plt.plot(df_sat.index,anom,label="Corrección")
plt.legend()
plt.xticks(ticks=df_sat.index[::5],rotation=90)
plt.title(f"Corrección de Anomalías de Temperatura en {nombre}")
```

```python
# Media cada 20 años
# Primero quitamos los datos que no tenemos registro
anom = anom.dropna()

division = [anom[i:i+20] for i in range (0,len(anom),20)]
bb_20 = []
for i in division:
    em = sum(i)/len(i)
    suma = 0
    while(suma<20):
        bb_20.append(em)
        suma+=1

# Quito los datos finales porque no nos interesan, no tenemos fechas que les corresponda
bb_20=bb_20[:len(anom)]
```

```python
# Plot de la media cada 20 años de las anomalías registradas
plt.figure(figsize=(8,3))
plt.plot(datak_interesantes,bb_20,label="Media cada 20 Años",color=sns.color_palette('Set2')[1])
plt.legend()
plt.title(f"Media cada 20 Años {nombre}")
```

```python
# Calculo los datos necesarios para poder realizar el gráfico de las temperaturas estimadas
mest=np.flip(mest)
tk=list(range(0,501,20))
logging_date = 2012
inv_all_years = list(range(logging_date-500,logging_date))

inv_all_anom = []
for i in mest:
    iter = 0
    while (iter<20):
        inv_all_anom.append(i)
        iter += 1
print(len(mest))
```

```python
# Gráfico de las anomalías estimadas y registradas
plt.figure(figsize=(8,3))
plt.plot(inv_all_years,inv_all_anom,label="Anomalías Estimadas", color=sns.color_palette('Set2')[4])
plt.plot(datak_interesantes,bb_20,label="Media cada 20 Años",color=sns.color_palette('Set2')[1])
plt.plot()
plt.legend(loc="best")
plt.title(f"Anomalías Estimadas {nombre}")
```

```python
# Gráfico con mismas dimensiones a el del artículo
plt.figure(figsize=(8,3))
plt.xlim(1500,2020)
plt.ylim(-2,2)
plt.plot(inv_all_years,inv_all_anom,label="Anomalías Estimadas", color=sns.color_palette('Set2')[4])
plt.plot()
plt.legend(loc="best")
plt.title(f"Anomalías Estimadas {nombre}")
```

### 06.3 - Media de anomalías estimadas en los *boreholes* 1501,1504,1505,DDH2457

```python
# Leo los datos estimados
mest=np.flip(mest)
mest_1501=np.genfromtxt("../resultados/mest.csv",delimiter=",")
mest_1504=np.genfromtxt("../resultados/mest_1504.csv",delimiter=",")
mest_1505=np.genfromtxt("../resultados/mest_1505.csv",delimiter=",")
mest_DDH2457=np.genfromtxt("../resultados/mest_DDH2457.csv",delimiter=",")
```

```python
# El logging date de todos es el mismo: 2015
# Asi que se hace la media directamente 
# No se tiene en cuenta el borehole DDH2457 

data = {"1501":mest,"1504":mest_1504,"1505":mest_1505,"DDH2457":mest_DDH2457}
df_mest_media = pd.DataFrame(data=data)

media_zonal = df_mest_media.mean(axis="columns")
```

```python
# Calculo los datos necesarios para poder realizar el gráfico de las temperaturas estimadas
media_zonal=np.flip(media_zonal)

# Definir parámetros necesarios
tk=list(range(0,501,20))[1:]
logging_date = 2015
inv_all_years = list(range(logging_date-500,logging_date))

# Media cada 20 datos
inv_all_anom = []
for i in media_zonal:
    iter = 0
    while (iter<20):
        inv_all_anom.append(i)
        iter += 1
```

```python
# Plot de las anomalías de los boreholes 1501, 1504 y 1505
plt.figure(figsize=(8,3))
plt.xlim(1500,2020)
plt.ylim(-2,3)
plt.plot(inv_all_years,inv_all_anom,label="Anomalías Estimadas", color=sns.color_palette('Set2')[4])
plt.plot()
plt.legend(loc="best")
plt.title(f"Anomalías Estimadas de los boreholes 1501, 1504, 1505 y DDH2457")
```
