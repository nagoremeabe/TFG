```python
# Importar librerías necesarias
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
```

# Step-Function

Perturbaciones de temperatura debidas al cambio espontáneo de 1 K o 1°C hace 10, 30, 100 y 300 años en la superficie de la Tierra.

## Cálculo

```python
# Definimos los parámetros
alpha = 31.6 #m**2 yr**-1**
tk = np.arange(10,310,10)
Tk = np.full(len(tk),1)
zj = np.arange(0,525,25)
```

```python
filas = len(zj)
columnas = len(tk)
```

```python
def Tt(zj,año):
    Tt_index = []
    for j in range(len(zj)):
        em= math.erfc(zj[j] / (2*math.sqrt(alpha*tk[np.where(tk==año)]))) #el tiempo en el que la temp crece
        Tt_index.append(em)
    return Tt_index
```

```python
Tt_10 = Tt(zj,10)
Tt_30 = Tt(zj,30)
Tt_100 = Tt(zj,100)
Tt_300 = Tt(zj,300)
```

## Plot

```python
# Plot de las anomalias

plt.figure(figsize=(3, 5))
plt.ylim(0,500,25)


plt.plot(Tt_10,zj,"aqua",label="10 años b.p")
plt.plot(Tt_30,zj,"dodgerblue",label="30 años b.p.")
plt.plot(Tt_100,zj,"blue",label="100 años b.p.")
plt.plot(Tt_300,zj,"navy",label="300 años b.p.")
plt.xlabel("\n Anomalías de Temperatura (K o °C)")
plt.ylabel("\n Profundidad (m)") 

plt.rcParams['xtick.bottom'] = plt.rcParams['xtick.labelbottom'] = False
plt.rcParams['xtick.top'] = plt.rcParams['xtick.labeltop'] = True

plt.legend(loc="lower right")

plt.gca().invert_yaxis()
plt.show()
```
